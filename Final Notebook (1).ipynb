# Intro
In order to determine the best neighborhood in Pittsburgh we decided to look at three metrics: Community, Education, and Safety.  We believe that each of these factors are essential in determining the best quality of life a person would have if they were to live there.  Having a good community with solid education in a safe environment is the dream scenario for where to live.  The dataset used for Community takes a look at the amount of family households in each neighborhood.  The Education dataset takes a look at the amount of students enrolled in school in each neighborhood.  The Safety dataset takes a look at the amount of arrests over a period of time for each neighborhood.  When combining all of these metrics we feel determining the best neighborhood would make itself apparent.

# Metric 1 - Community

```python
import pandas as pd

%matplotlib inline

import matplotlib.pyplot as plt

hhtype = pd.read_csv('HHDATA.csv')

pop_col = hhtype[['Neighborhood','Estimate; Total:']]

sort_n = pop_col.sort_values(by='Estimate; Total:', ascending=False)
top_10 = sort_n.head(10)

top_10.plot(kind='bar', x='Neighborhood', y='Estimate; Total:', legend=False)
plt.title('Top 10 Neighborhoods population')
plt.xlabel('Neighborhood')
plt.ylabel('Total Pop')
plt.show()

fam_col = hhtype[['Neighborhood','Estimate; Family households:']]

sort_f = fam_col.sort_values(by='Estimate; Family households:', ascending=False)
top_10fam = sort_f.head(10)

top_10fam.plot(kind='bar', x='Neighborhood', y='Estimate; Family households:', legend=False)
plt.title('Top 10 Neighborhoods by Family')
plt.xlabel('Neighborhood')
plt.ylabel('Total Families')
plt.show()
```

In this data we are measuring the household population of all total household types and then looking at specifically the family type of households. We have restricted the results to just the top 10 because we are only looking for the highest number of family households to determine our best community. Because Squirrel Hill South has a high number of family type households we think it has the strongest community. In a close second place is Brookline as it does not fall far behind in terms of family households. All in all both of these neighborhoods have strong communities because of the high number of families. Families encourage people to meet and bond with each other as their children 

# Metric 2 - Education

# Education

## Processing Raw Data


```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

df = pd.read_csv("schools.csv")

columns = df[['Neighborhood', 'Estimate; Enrolled in school:' ]]
sortBy = columns.sort_values(by='Estimate; Enrolled in school:', ascending=False)
number15 = sortBy.head(15)

number15.plot(kind='bar', x='Neighborhood', y='Estimate; Enrolled in school:', legend=False)
plt.title('Top 15 neighborhoods with highest education enrollment')
plt.xlabel('Neighborhood')
plt.ylabel('Students Enrolled')
plt.show()
```

## Data Overview

This data represents the total number of enrolled students in Pittsburgh's school system. By graphing the top 15 most enrolled schooling districts, we can draw assumptions about the safety and security of those neighborhoods.

We can clearly see that North Oakland has the highest student population out of every neighborhood in Pittsburgh. When combining both Squirrel Hill North and South, as well as North and Central Oakland, we see that the neighnborhoods with the highest student population are: Oakland, Squirrel Hill, and Shadyside.

## Conclusion

As with others, we see that Squirrel Hill is once again one of the more "successful" and populous neighborhoods of Pittsqburgh. Additionally, Shadyside and Oakland make strong cases for their schooling systems, based on enrollment.

# Metric 3 - Safety

## Intro
I decided to look at the safety factor in determining the best neighborhood in Pittsburgh.  This was done using a dataset that looked at the amount of arrests in each neighborhood.  I believe that the less arrests, the better environment it is in terms of safety, making it a good choice to live in.  To do this I was able to take the amount of arrests and compare them with the amount in each neighborhood.  The neighborhoods with the lowest arrest totals are the ones that are the most safe.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_set = pd.read_csv('https://data.wprdc.org/datastore/dump/e03a89dd-134a-4ee8-a2bd-62c40aeebc6f')

data_set.head()
```

```python
select_columns = data_set[['INCIDENTNEIGHBORHOOD']]
select_columns.head()
```

```python
neighborhood_counts = data_set['INCIDENTNEIGHBORHOOD'].value_counts()

print(neighborhood_counts)
```

```python
neighborhood_counts = data_set['INCIDENTNEIGHBORHOOD'].value_counts()

print(neighborhood_counts.head(10))
```

Looked at the neighborhoods with the most arrests, just to see what I was working with.

```python
neighborhood_counts = data_set['INCIDENTNEIGHBORHOOD'].value_counts()
bottom_10_neighborhoods = neighborhood_counts.nsmallest(10)
print(bottom_10_neighborhoods)
```

Organized the output by the 10 neighborhoods with the least arrests.

```python
neighborhood_counts = data_set['INCIDENTNEIGHBORHOOD'].value_counts()
bottom_50_neighborhoods = neighborhood_counts.nsmallest(50)
print(bottom_50_neighborhoods)
```

Noticed that none of the neighborhoods from the other metrics were represented in the bottom 10, so I increased it to the bottom 50 to look for when the first overlapping neighborhood.

```python
neighborhood_counts = data_set['INCIDENTNEIGHBORHOOD'].value_counts()
neighborhoods_at_least_100 = neighborhood_counts[neighborhood_counts >= 100]
bottom_15_neighborhoods = neighborhoods_at_least_100.nsmallest(15)
print(bottom_15_neighborhoods)
```

I narrowed down the lowest set by those with more than 100 arrests to try and account for population.

```python
neighborhood_counts = data_set['INCIDENTNEIGHBORHOOD'].value_counts()
neighborhoods_at_least_100 = neighborhood_counts[neighborhood_counts >= 100]
bottom_10_neighborhoods = neighborhoods_at_least_100.nsmallest(10)
bottom_10_neighborhoods.plot(kind='bar')
plt.title('Bottom 10 Neighborhoods with at least 100 Occurrences')
plt.xlabel('Neighborhood')
plt.ylabel('Occurrences')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Charted the bottom 10 neighborhoods with arrests >100.  When doing this Hays in the best neighborhood.

```python
specific_neighborhoods = ['Squirrel Hill South', 'Brookline', 'Carrick', 'Greenfield', 'Beechview', 'Brighton Heights', 'Squirrel Hill North', 'Shadyside', 'Mount Washington', 'Highland Park', 'North Oakland', 'Central Oakland', 'Bluff', 'South Side Flats'] 

specific_neighborhood_data = data_set[data_set['INCIDENTNEIGHBORHOOD'].isin(specific_neighborhoods)]

neighborhood_counts = specific_neighborhood_data['INCIDENTNEIGHBORHOOD'].value_counts()

neighborhood_counts.plot(kind='bar')
plt.title('Occurrences in Specific Neighborhoods')
plt.xlabel('Neighborhood')
plt.ylabel('Occurrences')
plt.xticks(rotation=70, ha='right')
plt.tight_layout()
plt.show()
```

Looked at the specfic neighborhoods from the other metrics' top 15 in order to get a better estimate of the overall best neighborhood.  When done this way, Squirrel Hill North was the best.

## Conclusion
In this dataset we are measuring the amount of arrests that happened in each neighborhood in Pittsburgh.  We first looked at the top 10 neighborhoods in terms of arrests, then the bottom 10. However, compared to the other metrics, this neighborhood as well as many others did not match up.  So, in order to better compare between metrics we took the top 10-15 neighborhoods from the other metrics and analyzed their amount of arrests.
