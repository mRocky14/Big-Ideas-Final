Oh yeah, and let's import pandas for this notebook.

```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt

df = pd.read_csv("schools.csv")

columns = df[['Neighborhood', 'Estimate; Enrolled in school:' ]]
sortBy = columns.sort_values(by='Estimate; Enrolled in school:', ascending=False)
number15 = sortBy.head(15)

number15.plot(kind='bar', x='Neighborhood', y='Estimate; Enrolled in school:', legend=False)
plt.title('Top 15 neighborhoods with highest education enrollment')
plt.xlabel('Neighborhood')
plt.ylabel('Students Enrolled')
plt.show()
```

----

## Data Overview

This data represents the total number of enrolled students in Pittsburgh's school system. By graphing the top 15 most enrolled schooling districts, we can draw assumptions about the safety and security of those neighborhoods.

We can clearly see that North Oakland has the highest student population out of every neighborhood in Pittsburgh. When combining both Squirrel Hill North and South, as well as North and Central Oakland, we see that the neighnborhoods with the highest student population are: Oakland, Squirrel Hill, and Shadyside.

## Conclusion

As with others, we see that Squirrel Hill is once again one of the more "successful" and populous neighborhoods of Pittsqburgh. Additionally, Shadyside and Oakland make strong cases for their schooling systems, based on enrollment.
